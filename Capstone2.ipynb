**1. The percentage of road accidents during all the years.**

```python
data=[df["State/UT-Wise Total Number of Road Accidents during - 2013"].sum(),
      df["State/UT-Wise Total Number of Road Accidents during - 2014"].sum(),
      df["State/UT-Wise Total Number of Road Accidents during - 2015"].sum(),
      df["State/UT-Wise Total Number of Road Accidents during - 2016"].sum()]

labels=["State/UT-Wise Total Number of Road Accidents during - 2013",
        "State/UT-Wise Total Number of Road Accidents during - 2014",
        "State/UT-Wise Total Number of Road Accidents during - 2015",
        "State/UT-Wise Total Number of Road Accidents during - 2016"]

plt.title("The percentage of road accidents during all the years")
plt.pie(data, labels=labels, autopct="%1.1f%%",colors=sns.color_palette('husl'))
plt.show()
```

```python
data
```

**2. Mean Accidents per 1L population for each year.**

```python
data=[df['Total Number of Accidents Per Lakh Population - 2013'].mean(),
df['Total Number of Accidents Per Lakh Population - 2014'].mean(),
df['Total Number of Accidents Per Lakh Population - 2015'].mean(),
df['Total Number of Accidents Per Lakh Population - 2016'].mean()]

labels=["2013","2014","2015","2016"]

plt.title("Mean Accidents per 1L population for each year.")
plt.pie(data,labels=labels,autopct="%1.1f%%",colors=sns.color_palette('Set1'))
plt.show()
```

**3. The highest number of accident states and least number of accident states.**

```python
for i in range(len(df)):
    df.loc[i, "Total Accident State/UT wise"] = df.iloc[i, 2:6].sum()
```

```python
df.head(5)
```

```python
df_sorted=df.sort_values(by='Total Accident State/UT wise',ascending=False)
print("highest number of accident states --->>>" ,df_sorted[['States/UTs', 'Total Accident State/UT wise']].head(1))
print("least number of accident states --->>>",df_sorted[['States/UTs', 'Total Accident State/UT wise']].tail(1))
```

```python
plt.figure(figsize=(15,10))
plt.xticks(rotation=90)
plt.title("Total Number of accident in each state",fontsize=20)
sns.barplot(x='States/UTs',y='Total Accident State/UT wise',data=df_sorted)
plt.show()
```

```python
Min_values=df.iloc[:,10:14].min()
Max_values=df.iloc[:,10:14].max()

plt.figure(figsize=(10, 6))
barplot=sns.barplot(x=Min_values.index, y=Min_values.values, color='skyblue', label='Min')
sns.barplot(x=Max_values.index, y=Max_values.values, color='salmon', label='Max')

plt.xlabel('Year')
plt.ylabel('Total Accidents per Lakh Population')
plt.title('Min and Max Accidents per Lakh Population (2013-2016)')

x_labels = ['2013', '2014', '2015', '2016']  # Replace with your desired labels
barplot.set_xticklabels(x_labels)

plt.legend()

plt.show()
```
